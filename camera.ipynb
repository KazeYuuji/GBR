```python
# ====== 1Ô∏è‚É£ Import Libraries ======
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import pickle

# ====== 2Ô∏è‚É£ Contoh Dataset (bisa kamu ganti dengan dataset kamu) ======
# Silakan ganti file ini dengan dataset kamera milikmu
data = {
    "brand": ["Sony", "Canon", "Sony", "Nikon", "Fuji", "Canon", "Nikon", "Fuji"],
    "megapixel": [24, 20, 42, 45, 26, 30, 21, 16],
    "weight": [450, 520, 650, 720, 390, 500, 680, 410],
    "width": [127, 139, 126, 146, 121, 132, 147, 118],
    "height": [94, 98, 96, 115, 83, 92, 114, 82],
    "depth": [60, 78, 57, 62, 53, 74, 62, 48],
    "price": [900, 800, 2300, 3000, 1100, 900, 2800, 850]
}

df = pd.DataFrame(data)

# ====== 3Ô∏è‚É£ Encode Brand ======
encoder = LabelEncoder()
df["brand_id"] = encoder.fit_transform(df["brand"])

# Dimensi jadi satu angka (volume)
df["dimension_volume"] = df["width"] * df["height"] * df["depth"]

# ====== 4Ô∏è‚É£ Dataset Split ======
X = df[["brand_id", "megapixel", "weight", "dimension_volume"]]
y = df["price"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# ====== 5Ô∏è‚É£ Train Gradient Boosting Regressor ======
model = GradientBoostingRegressor(
    n_estimators=350,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.9,
    random_state=42
)

model.fit(X_train, y_train)

# ====== 6Ô∏è‚É£ Evaluation ======
y_pred = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("üìå RMSE:", rmse)
print("üìå R¬≤ Score:", r2)

# ====== 7Ô∏è‚É£ Save Model ======
with open("GBR_camera_model.pkl", "wb") as f:
    pickle.dump({
        "model": model,
        "encoder": encoder
    }, f)

print("\n‚úî Model saved to: GBR_camera_model.pkl")

# ====== 8Ô∏è‚É£ Tes Prediksi Contoh ======
test_sample = [[encoder.transform(["Sony"])[0], 24, 450, 127 * 94 * 60]]
test_pred = model.predict(test_sample)[0]
print(f"\nüéØ Test Prediction Example: ${round(float(test_pred),2)}")
```
